```python
import os
from langfuse.openai import OpenAI # Use the Langfuse wrapper!
from langfuse import observe
from dotenv import load_dotenv
load_dotenv()
```

```python
# Initialize the client for local Ollama
client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama" 
)
```

```python
@observe() # Simply use the flattened import
def ask_ai(question):
    response = client.chat.completions.create(
        model="gemma4-pro:latest",
        messages=[
            {"role": "system", "content": "You are a helpful AI assistant."},
            {"role": "user", "content": question}
        ]
    )
    return response.choices[0].message.content
```

```python
if __name__ == "__main__":
    print("Connecting to Gemma via Ollama...")
    # Type exit to break the chat
    while True:
        question = input("\nWhat is your Question: ")

        if question.lower() == 'exit':
            break
        
        print(f"\n Gemma's Reply: {ask_ai(question)}")



    # answer = ask_ai("Explain the benefit of Langfuse v4 in one sentence.")
    # print(f"\nGemma says: {answer}")
```
